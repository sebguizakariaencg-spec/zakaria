```python
# =============================================================================
# TP PARTIE 1 â€” STATISTIQUES ET LOI NORMALE EN FINANCE
# ZAKARIA SEBGUI ET SAAD EDDINE MOUALI
# Analyse risque portefeuille et calcul VaR
# =============================================================================
# ZAKARIA SEBGUI & SAAD EDDINE MOUALI

from IPython.display import Image, display

# For a publicly accessible image URL
display(Image(url='data:image/jpeg;base64,/9j/4AAQSkZJRgABAQAAAQABAAD/2wCEAAkGBxALEBAQEBAKEBAJDQoNDQkJDRsIFQ4WIB0iIiAdHx8kKDQsJCYxJx8fLTItMSwtOjEwIys0OD8uQDQvMC0BCgoKDg0OFRAOFTcZFhkrKzcrNzctLzcrKysxNysrNysrMS0rLS0tLSs3NS83Ny0rKysrLTcrKzcrLSsrLS0rK//AABEIAMgAyAMBIgACEQEDEQH/xAAcAAABBQEBAQAAAAAAAAAAAAAEAAMFBgcCAQj/xAA9EAABBAEDAgQDBgUCBAcAAAABAAIDEQQFEiExQQYTIlFhcYEHFDJCUpEjYqGxweHwFUOC0SQzRGNysvH/xAAZAQADAQEBAAAAAAAAAAAAAAABAgMABAX/xAAlEQACAgICAgIDAAMAAAAAAAAAAQIRAyESMQRBE1EiMmEjM3H/2gAMAwEAAhEDEQA/AJDFnArvdfBWDSHwiVvmE0ao9lUI5RJ6mkmuvwRmLmOaet7frS4qraJdGwxkUK6UKXSr/hvVzkNDCLLRy5WBdcZKStFUxueTY0n2BUYzPLQTXU9U/qUgLS0ckdhyoqWQMY2zZNelOYn4Zw5oPunlB6Zlbn7eynEDCSSSWCJJcklKz7LAOl44LyyvQbWMDOeQaoJ4N46cpt0VutP9kXQisFLzdem/ZPOjDhyBaYGN6rPugtV1I4x5Bo9Ci0jRbIvXWc9ht7Kv5MwPH5vgpLMyfPDibUXtA5pSb2Md7ywDnqnGAuBdxXRByu+pRAa9rAegeQgwoMmc0xAhoBJIL75KIhIMQY3kyEFx/SouSYAAAjlPYOUQ5jLrzHAG0E90EMytOMTS4lpLa4+C8TWrueJCAeOG8pJjWUPToDvIDttg9OQjmxvgJLnsO/sFBY07muDQbHCm5IxY9RJDeA7lc8m0/wCEiw6NmugcHNP4qsHhWaLWHOPVUPCyHH0urg9hVK2adp7pGWK9X4Se6TDJ8mhokvFK0Bzi7l1qBzcvc412uk8/Fe3c03bPboo7y7F9122MWHw/K0cuPq7KzNeSqt4fw91uPBrgq0YzaaB1RMeh5Xu8roBeV3WAIuA6mkt4Vf8AEmtQwxvp4dJG5rfKjPIPv9FSH+OZYt1Ovbw0FtJlFsJrKbF7u1LP8P7Q9sY8xoe8H1Fg2gBWPRvFUWbJ5bRRDbMpcA274AQcWjEy7JaHVzafJTHkeq0+Qi6Fi3uxhmU1xoXa8zcNk4p4sBex4+02npBwa9kHXo0W62VbVMSONtMBtQmVgujALiKd7dlLarkua/y6Fm0JqML5GNYDb3fGlOdegxt9jGjGFjyZA02307huT2o5bS3a0cNJICi24j473dYyAa5Sll3ODQeDQtTt9DHrcZzzuAJr60pXDx4ZAxj2uD27nGRvHKkWYfkQFzCHWL91XoMzl3v6gnqjHufKXkC7LON3ukhWWSb+K8U+RqM/xZLIvp7qXx8sSna7qAACFDYLzTmkc9bPVSOnEB3JF8DcRaWatEyYxHODg1xHP4XjlT+napO07Gud/Dumt5Qel6E/IjdO1rdsV/mon6J3SWPbLbQTuvhTjjd30FInpNVe6H8Hr5DnlROBPe5rvorZBDFI3aQGv7g8qPzsXHhZICBvO7ae99qXYuhjnF3QwucHfRFaJqcjquyEDC7y8fk3u6tKP0PMjaNtBDkYskdnlQ/iXVxgR7nEjzNzIzXG6iQCe3RPMzyxxutvZU77Stcb93DNrXCR/c0W8dQnTV0ZGda1qj5pHPIawyOc6o3b+vXlRr8kuqybC7wMZ2W7vtaRucpuLT42cAD68qjmolIYnPZAnJIFWa4R+JqZjqi4Vz1KknaPHJ2/bhROp6NJi+rksv8AF7LRyJhnilDZrvg7xeMhkcUxBlO1gk6blb8zJbC0uPbssB8I6gyPJhMnmeWx1u2c8rYfvMebYD+vS0sqRIUXiQSProEbJrbGD1Ai+nxVextKayX1PG0G0Zq8AkaSHN9IO2klg2B5maySXzKUfNlOaS4E3fp7pmYmNoJ5KPhwfPjBFBx/UoybsKBcaeSYuYAXGXkoaTHLXFrhRae6MwA7GmINbhbb6hOlnnz+o9wSQk5Bqyx6K3/w7r54PVVQYZ8xzgOA5ysGVnjHhLGd+FA4+eRYrh12VZzizUCPd6+lUkvct/NgfNJQbMUXDeHvPA9Q+SkRA0EOaaLRRjKBbgvx2tkcCA8W0O43fJTWgQDIdRPL+loN10ITOhGWMAB7gx/DmA1YVhawRtLmAA1whGYRhYOhPHARrQ4gCuyZXRWNdEI2ecvLy43fbhNZepuaacCT+rqjMuNzCeOCvYMITt52/VNF+hOOwTGzw9pv9k7izHeNt8lTOFokMdXso/FdxQRRS8AUOiLX0Moh0mONo3HlwHwVH+0OBjMc2HOcxzXNLeQ35q7Ty7+4odOVEa1p4yIZWl1ebG9u/wDFXCEbTtjcb6Mv8LODmPBIFEuJ6KSjnje4hrmkj2NqH03BdG3IaRu5ZQYaDhz/AEXGMyRjxQjaLHDBSedM6sVpJUWNmZHCR5hAB6Wn/EE7XYbnMLXNNct5UPrGFKdhb+FzWkGrUhi6e9+JMHiOnRuILPTyORwtB0Nkt2VLTx50jGEuAcereDXdaNornMLQA+gGizyoHwboHl5DHSFpAie4sI6ey0dsccYsBqec7WjjeNr9hsQB3U8r2XGsEC1zCQ5xcLKKe8AWubfsPFFazHbBRs0m8PWHAhosKYynR1ucAQgDmYrDYbz8E/8A0TiM5shB3c27klc6fOQ7ryUtUmbK0OZYCa0iVsbreCR2rlScWNVEhlQPcbs0V7iaYW+oknd0CLmzGygCMH9qXYl2sBPUJopezUgDJqPqACOxSTWpSk+sg8VXZJSm9iN7A9V0R8whvcWx2NvYBGaZpBi27WkBhtXTIzYI7B2mlSda1uZjyI62nsF0tRs1UWP+GAS91fy9UHJqMYNNBJVaxzLMQXF3KnsWBsYs8lFySQ8Uhn7rLKS43R6BSAxGCPaAQ6hz3tcR6kHHaAiwS5BSrtD8SOdgudXrdQ+KfmhcQAD07oxoa7i238165ob3RujUgKDT3d3mkVNgtLC3ceQQuHSj3XoyAEvIGkZ8/DdhyTRvHNNLXfqA/wD1ReRktYQSL77RwtK1OFmXG4FrNxY5rZC3lqyzNxiX9SNh5aOLSvZ2Yp8rom/+IiWFoaOIT/5l19FLYE9xnpyDx7qrxQ+ZXLmV2j4v5q1+HsZsjmseLAAcW/LkLXspPS2SPhvQRsD5HHgbWVxx1UtkYTCKaXWF5mgtYdpraOnRRum5hLrJukbPOnkcpWyw6fjNjaARyeqey8ZpY4Ac0aTWFKZUW5ijOcuh4RTVlJy4pRYcOLKZg02R4BDY/V+oqfzm8m/io7ziygC4V0rhVT+yb7HmYFQuDmi2muOVCkuhJ4Fc1auuAzdCSe9k3yqd4kx3vsM45PRXcbijKQTp0ORMC6MMo31O1PjByvzhvp9nWo/RpJsZm0vcKRb9SkP/ADCffnqueXFasVzS7HM+Nz2BrhW7gFJMSZ3S9x/rS9SwlGtito93Of7lIYgJs19URjRuHVP/AHcO63/ZPa9FRzHgY2uh+SLbjB9jgcIeKIRqD8Qam+E2CW17Gk+LHyexW+KsOhsSlgbwCfX1QGqSZTJCGk7T3U54PfHPjGZ57uHK8j1Ns0wiIjEd0XlVceLH+Tl2igQ5OXFkAl5LdwsEq7QZzZa9QJIFttQf2k6A/CZ94iedjiAQONqrHg/UA7Ij3F3tyeqEoNqxZNPo0k/VPRxX1RkIHsm8vLjhHqIH8vdQjFt0g8Ets4EAWYa7GcfJmZ12SOoHjjstBxNdZJNHGA0CVzm2Tuvgn/Cr3j/EhcA8yRxTFxbC6Q7fN/l/30VHhaWx8eVReitwTOPSgO7ldPA2G7IdI/sxjmB57u9lQYYZqf5lRR47d0uQSJBXsK6krVfs61XEzsUNxSWHFDRLjyH1sPufe/dbHj3srmzpqkcZN2Wu4PIrohIMTYTwogag9uRM2Qk7MrJaR0/MVO4+Q13Qn/qTPE/RLNj+Nq33sn9HZTLI6qQ22h9OmY5gDXNJbW4DsiwueUaezJ6IDPi9ZBQRxx1oKX1th6hQhJ7nqmYj7C48lzG7QeEJIzd15tctjH6kQ2qRUn0zURWr01oA7/ipRWOWWXW7aOpcpmeBzyeLHsq3qEjoyWua4Ns8AdVGUd2yMoslMXIY/dtc0hgJB90lW4XEFzg3a34GiUklIU0sxcCguTGeykYqofILvaCq0dlES+EkE+wKo3iounaSCbiPIC0t0N/VZl4inONmGJwpkwd6j0Vsbpm1VP2Tei58YwQyPgsHqCZbksBaQebFoXwzi+aJw3tdUvYcOTeAWE8jsqZHvROC1TJnxxlCbTJAXj0bOD3WUeH5i3IgLQT/ABG2B7K9/anpz2wQlocGvA3sbwLVQ8O44hY6VwAcQWsDuDXc/wC/irJaGwYvlyKKNBzvFu0VH6Q07HEj1AqrZut+by5xJdfJN2q9n5xDj8gL9x2KjW5l9zaEGl0dnnKEYKMUWGXWXwywSM6wyEgE1fBH+Uf4jxf+MNfkbnfw2s8r+QVyK+dqmmXe9g5Ibuc6lO6JrYwhKH+qORjjt/mA4/fp+yWcrZz48KeKyL0HT5M6T7uXygFzi9u7gV3pWrOidoLYX4bnRTSlzjP+NxYPf4H2Va8K6kYsve8gfeHPbIT6atSfifU/vcz3A+hp8uMezR/u/qg9G8XEpyd9ILdrT8qQzSCNsk72mTyhsbdcmvjV/VP/APGzH3vqqxHPTAfaUX+xQ+TlHn6qkZaN5n+xL+IvWna4/cHtcQXG9zTt4C07w1rLc6PkgSR1uHTd8QsD0zMoAfAC/YK6+FdW8ieJ18AgObfAaeFOX5L+npvDHJgVdpGpay30WOyr7YC/m1Z5gJWkUoPIxyDQNLlZ49bAMiOuiaMj2HgWE+7FcbFnlc+QWDk3SUNHbZXn8tLmbGEo5aD8040vpGRgVZWtjJWQMuhscbql6rACw8FJakZxR5jSuqiOB3RIemWxSV+E17pGN9dCnaByQ46YjhZz9q0Y2RyDhwPULQBA/vazf7W3lvls5rhHH+wJy0U/StcycQOdG8i+t8qXg8Y54YZQG7WGi+rTbdLA04SgcvdyVaNJ00DSJGlrbcyQ2fkrykSRWJvFOVqHEztzR0aBSH1HJLAxh4LWin9avqClpeFWLJJxbSEPrt3uIJDqv+Up3O40d3gRpzmRuU7/AE70owPon4Eo54sccj/6qLc6iQRyD1CSIPM3RI4RNl1+zK90RMP2Q8Qpte3J+ad32B+xStnThjWNRBoI/XXbr9FISPsIZnFn3sArrctZsUFBOvZ20kxyAclu2Svl1/oUBPIj8N9OPxDh7qMymU7aOhIrvwmTObyoXxkiQwjVH26d1Y9JebBJ68j4quY9CvoA1T+muo/qea4HRqy7PR8XWjdfB2d96xWE/iYDG8/Ef6UmNQZteeqA+y4n7tLZv+Pf12hTWts2eoDqp5o60eT5C45pJEQ6Yt5LTXumpMgEdPZOOmseroOyjJcnkgdFy2yPIkJJd9V0pDuc4dwPmgZZnDogcnJffX2Rts3yUWFzXiqcCElXcfKf2ceUlg/Kafpe50MZJ5LG2eqK2H+X9kLooIgiB6hgCOXdROxnyz7N/ZZn9rWhS5joPLZZog7AtSQ+S0nkVYv8XKyVGMP1LHkxNPjgkFO8wClfHYgGlVQsw2nPGehHMijIa3c2UE7fZT33Jv3URkCxDs+tI0mHoyfwrBGcHKL235W+ife+FVcuYbyHcslr40rrmaBkYmLO7c1sQc50jelttZ/qL6JaQeAKI7oM9Twmo4m/6DSwGJx9jw4dVD5TNslfJTEWRu9J6i9pPceyi9S4ePklRvKUXG19hEbk47ix7oZh4RBNtB9uEGNB2jryzsa6uC4jd8VwTX1S3HYBbqBvbfFry7H/AMSUAnULqchsg25vwJXYdSe07EOQ+/yxkE/H4I9EZu0kdQHv09lMYOQGim8uP5Ixf7lSeNAIZPMLYg+m/l4iFV+6AngkjkL4GOMBIIBouPv07WtF30dWFtM137KnHyJmEguEoea56j/RXTLgErCD1ANKgfZY/wBczQCA6KJ9Hsb/ANVoM0b/AMpCeSvR53mqs0im5AG5zTYq0C8MZ1IV6OEHA7mMJPdCyaLEf+W0rn+JnKVB+OXi2HrZpCjDJ5I7rQGaXEBXlgfJNy6NC7s8fIrfEwFAOJsuhykr4dDh/nSQ+NmCtKcBE0e25Gbx7hUfLynsPpeQOOEBJqsrgQJHg9irSyJOg0aRabn5CzGTWM2Oqldz17onF8QZgNSOBHuQh8qDx9l8LWltHmj3T/liqpUuPX5dtU2wjoPEzzdsbxSKyIzCPGeLu07MDR/6eY/tyvnY5JZ6ZG7mi63dvkey33UPEXmwTNMYp0MwPPwKwLKIk6uILeKA3I8kzt8RvixvK8o0RvaDW15G8WhZMCTIeNm13pHN8FF4zY2k7jIb9+n7IzAkbC+2bDfblyF/R0zx/IjjB8JZctcQNHvI/wD7Bd5nh2bEIbK6Gn2QY3F/9KVmwtTAI3H/AKAdpKgdc1MSzu5NMprR+JKpWzY8ajL8iJkwJBwNjhZohwauodMkN0YzdekHlFRTMJ6j+ycfO1ptpv8AomOlYsb2RMmmyg0G77B9UfIUrpWDJEwAhrSSXep3f6If7y6B5rlkhLg0/wCET95Eg9LyP/bk/wC6z2TjhxXfsJ1GN0ke3d8XFvO4/FR+n6g7FPlncWE8+rf+xXTp3jrf90O4WenBWWugzaTuGmbL9kwuSYjkeTDTvcErS1mn2Ns9M5/SzGb/AHWlKh5fmu8zPUl4laxynqS8SQMepJJImMw1l5DgASBtCEZktZVDpVk8papNuDCTzRBQMjxVry88msjEcqYfJni/muHZN9So7eO/UlPMI6HtfVT5Ng5sNbkV9UTBNZN2EAw2Pl9U4Zqr58orJRrCcgkgi+HhwrosgymkOcCKLXOBC1YS2VT/ABD4fkMrpImOeyb1FsfqLT34V8GTbTOzxJpNxb7KkSiYX7OensE8/EdF1Y5p9ngtTbo7XXZ6kYtbQ75gkFO5+aj8nHcw20kg+/rROyl4b7Baxprkt9gkbr62Cnmkj8w+qT4yeq7x8cvc1o/O5o9XZYkrR2HAijVLgQkdOQpubSsRvXIjafbeJP6KOkiiafRIZB+pjSz+6ClY2mNNc4dQV213wCejaCaBd9eFongXwNh6nCZpZMhzo5Cx8EZEYHQ8nryD8EyBkahHky0fZJhOiwjK4UcyRz2j+UcD+tq7prFx2QMbGxoayFrWMY3gNA6BOJzyMk+cnJnSi9T1Y4rw0sJDwTuBpSahfE+nHIjBaaMRv6IPS0KqvZ5D4kY7qx44+adh8Qwv/WKvqFWRFsoddoFroEDsBXboofKxqRa49Zhd0cfqKSVMbKb44u+EkPmYOKK5qj6I9jaFa83R6cHlE6ww+kjuhWGwOq5vJX5sjPsLYAT8qXEgo2vAQOffhchxuv1dCoCjkcpBrsnonDn/ACh3wU0kcn4L1jXdTwtxoI+HBpNFGxyemmgXzZKj2R2evXsinPaO9dKHRGKYUiteMpiNgN2Gu72qVJI7s537q5eNzuMZogAOF9bVMfwV6OJfij2ML/xRGnZEgNbj+wXj5n/qP7UvZG3yk31D4hUBcvsURc6zud196XrgeOSfmbTmLtAeHbrr017pVawVtHLG0VIYuPYO33Jr2QaLxHlp4+CxXHV7C4YHX0WifZDn+XkT47jX3mNkjQe5b1/of6KkR5NDkf4Wg+BvCk4mgzZAyONrTIxu63vBHHHYc2mRTy1jWJps060l4vbRPBPbXj2hwIP5gQvbtJYJUsrC2vNnpYA6IM47ibH7HlTuvM2kGuH1z7FRfmt6WuOUalRWtWB7BzxyAaSREm0ci/n1SQpCFU1dtsYfcN5CiYpfV3r26JJJfI/dkp9hYomh0d/dd8N7XdJJKCFPQw3wfonWDkC+Skkh7McElriKT+PFvIvkAjjokkmih4og/Gr2iKUhoBjfC1jXGxfX+wKoLZBJ04P6HcFJJehjX4nXgm0+Po9IIXA4NpJJjsY6GjcAOrr4T7Yq6pJIFII4fGb4CKx4qFnj58JJIjRS5BX3hoI+FcnlfRWhZgysaCVvSaCJ3tRrkJJJ0jg8yblJJh9r20kljiPGOXQKSSwQXVMcTRkH8tFV52M0H+m5JJQy9lIdAueOA1po88dLSSSXOwH/2Q=='))

from IPython.display import Image, display

# For a publicly accessible image URL
display(Image(url='https://media.licdn.com/dms/image/v2/D4E03AQH8VtYK8TYgRw/profile-displayphoto-shrink_400_400/profile-displayphoto-shrink_400_400/0/1716369405233?e=1772668800&v=beta&t=rpXScbwIEzDjPzzgeZEE1aFJaBz9wlJMXmWM5dUsjCU'))




import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Configuration graphique
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 6)

print("="*80)
print("TP PARTIE 1 â€” STATISTIQUES ET LOI NORMALE EN FINANCE")
print("Analyse risque portefeuille et calcul VaR")
print("="*80)

# =============================================================================
# DONNÃ‰ES
# =============================================================================

# Rendements mensuels historiques (%)
rendements_A = np.array([
    1.2, 0.8, -0.5, 1.5, 0.9, 1.1, 0.7, 1.3, 1.0, 0.6, 1.4, 0.8,
    1.1, 0.9, -0.3, 1.2, 1.0, 1.5, 0.8, 1.3, 0.9, 1.1, 1.2, 1.0
])

rendements_B = np.array([
    4.5, -2.1, 6.2, -3.5, 5.8, 7.1, -1.8, 4.9, 3.2, -4.2, 8.5, -2.7,
    5.1, 6.8, -3.1, 7.3, 4.5, -2.9, 6.7, 5.3, -3.8, 7.9, 4.2, 5.5
])

capital = 500000
perte_max_toleree = 50000
taux_sans_risque = 3.0


# =============================================================================
# QUESTION 1.1 â€” STATISTIQUES DESCRIPTIVES
# =============================================================================

def calculer_stats_portefeuille(rendements, nom):
    """Calcule les statistiques descriptives d'un portefeuille"""

    moyenne_mensuelle = np.mean(rendements)
    ecart_type_mensuel = np.std(rendements, ddof=1)
    mediane = np.median(rendements)

    rendement_annuel = ((1 + moyenne_mensuelle/100)**12 - 1) * 100
    volatilite_annuelle = ecart_type_mensuel * np.sqrt(12)

    return {
        'nom': nom,
        'moyenne_mensuelle': moyenne_mensuelle,
        'ecart_type_mensuel': ecart_type_mensuel,
        'mediane': mediane,
        'rendement_annuel': rendement_annuel,
        'volatilite_annuelle': volatilite_annuelle
    }


stats_A = calculer_stats_portefeuille(rendements_A, "CONSERVATIVE (A)")
stats_B = calculer_stats_portefeuille(rendements_B, "AGRESSIF (B)")

print("\n" + "="*80)
print("QUESTION 1.1 â€” STATISTIQUES DESCRIPTIVES")
print("="*80)

for s in [stats_A, stats_B]:
    print(f"\nðŸ“Š PORTEFEUILLE {s['nom']}")
    print(f"â€¢ Rendement mensuel moyen : {s['moyenne_mensuelle']:.2f}%")
    print(f"â€¢ Ã‰cart-type mensuel : {s['ecart_type_mensuel']:.2f}%")
    print(f"â€¢ MÃ©diane : {s['mediane']:.2f}%")
    print(f"â€¢ Rendement annualisÃ© : {s['rendement_annuel']:.2f}%")
    print(f"â€¢ VolatilitÃ© annualisÃ©e : {s['volatilite_annuelle']:.2f}%")


# =============================================================================
# QUESTION 1.2 â€” VISUALISATION
# =============================================================================

print("\n" + "="*80)
print("QUESTION 1.2 â€” VISUALISATION DISTRIBUTIONS")
print("="*80)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogrammes
axes[0].hist(rendements_A, bins=10, alpha=0.6, color='green',
             edgecolor='black', label='Portefeuille A', density=True)
axes[0].hist(rendements_B, bins=10, alpha=0.6, color='red',
             edgecolor='black', label='Portefeuille B', density=True)

axes[0].axvline(stats_A['moyenne_mensuelle'], color='darkgreen',
                linestyle='--', linewidth=2)
axes[0].axvline(stats_B['moyenne_mensuelle'], color='darkred',
                linestyle='--', linewidth=2)

axes[0].set_title("Distributions rendements mensuels")
axes[0].legend()

# Boxplots
axes[1].boxplot([rendements_A, rendements_B],
                labels=['Portefeuille A', 'Portefeuille B'],
                patch_artist=True)

axes[1].set_title("Boxplots comparatifs")
axes[1].axhline(0, linestyle=':', color='black')

plt.tight_layout()
plt.show()


# =============================================================================
# QUESTION 1.3 â€” VALUE AT RISK
# =============================================================================

def calculer_var_portefeuille(stats_dict, capital, alpha=0.05):
    z_alpha = stats.norm.ppf(alpha)

    var_mensuelle_pct = stats_dict['moyenne_mensuelle'] + \
                        z_alpha * stats_dict['ecart_type_mensuel']

    var_annuelle_pct = stats_dict['rendement_annuel'] + \
                       z_alpha * stats_dict['volatilite_annuelle']

    return {
        'var_mensuelle_pct': var_mensuelle_pct,
        'var_annuelle_pct': var_annuelle_pct,
        'var_mensuelle_euros': capital * var_mensuelle_pct / 100,
        'var_annuelle_euros': capital * var_annuelle_pct / 100
    }


var_A = calculer_var_portefeuille(stats_A, capital)
var_B = calculer_var_portefeuille(stats_B, capital)

print("\n" + "="*80)
print("QUESTION 1.3 â€” VALUE AT RISK (VaR 95%)")
print("="*80)

print(f"\nCAPITAL INVESTI : â‚¬{capital:,.0f}")
print(f"Perte max tolÃ©rÃ©e : â‚¬{perte_max_toleree:,.0f}")

print("\nPortefeuille A")
print(f"VaR annuelle : â‚¬{var_A['var_annuelle_euros']:,.0f}")

print("\nPortefeuille B")
print(f"VaR annuelle : â‚¬{var_B['var_annuelle_euros']:,.0f}")


# Test normalitÃ©
stat_A, p_A = stats.shapiro(rendements_A)
stat_B, p_B = stats.shapiro(rendements_B)

print("\nTest Shapiro-Wilk")
print(f"A p-value : {p_A:.4f}")
print(f"B p-value : {p_B:.4f}")


# =============================================================================
# QUESTION 1.4 â€” RATIO SHARPE
# =============================================================================

sharpe_A = (stats_A['rendement_annuel'] - taux_sans_risque) / stats_A['volatilite_annuelle']
sharpe_B = (stats_B['rendement_annuel'] - taux_sans_risque) / stats_B['volatilite_annuelle']

print("\n" + "="*80)
print("QUESTION 1.4 â€” RATIO SHARPE")
print("="*80)

print(f"Sharpe A : {sharpe_A:.3f}")
print(f"Sharpe B : {sharpe_B:.3f}")

print("\nâœ“ FIN PARTIE 1")

# =============================================================================
# TP PARTIE 2 â€” THÃ‰ORÃˆME DE BAYES ET SCORING CRÃ‰DIT
# Mise Ã  jour probabilitÃ©s risque avec nouvelles informations
# =============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

print("="*80)
print("TP PARTIE 2 â€” THÃ‰ORÃˆME DE BAYES ET SCORING CRÃ‰DIT")
print("Mise Ã  jour probabilitÃ©s risque avec nouvelles informations")
print("="*80)

# =============================================================================
# DONNÃ‰ES
# =============================================================================

taux_defaut_base = 0.05

segments = {
    'Premium': {'proportion': 0.30, 'taux_defaut': 0.015},
    'Standard': {'proportion': 0.50, 'taux_defaut': 0.05},
    'Risque': {'proportion': 0.20, 'taux_defaut': 0.15}
}

evenements = {
    'Retard paiement': {
        'P(E|Defaut)': 0.80,
        'P(E|Non-defaut)': 0.10
    },
    'Decouvert >500â‚¬': {
        'P(E|Defaut)': 0.65,
        'P(E|Non-defaut)': 0.15
    },
    'Refus credit ailleurs': {
        'P(E|Defaut)': 0.55,
        'P(E|Non-defaut)': 0.08
    }
}

# =============================================================================
# QUESTION 2.1 â€” CALCUL BAYES MANUEL
# =============================================================================

print("\n" + "="*80)
print("QUESTION 2.1 â€” CALCUL BAYES MANUEL")
print("="*80)

prior = segments['Standard']['taux_defaut']
likelihood_defaut = evenements['Retard paiement']['P(E|Defaut)']
likelihood_non_defaut = evenements['Retard paiement']['P(E|Non-defaut)']

p_retard = likelihood_defaut * prior + likelihood_non_defaut * (1 - prior)
posterior = (likelihood_defaut * prior) / p_retard

print(f"\nPrior P(DÃ©faut) : {prior:.2%}")
print(f"P(DÃ©faut|Retard) : {posterior:.2%}")

facteur_multiplication = posterior / prior
print(f"Multiplication du risque : x{facteur_multiplication:.2f}")

# =============================================================================
# QUESTION 2.2 â€” MISE Ã€ JOUR SÃ‰QUENTIELLE
# =============================================================================

print("\n" + "="*80)
print("QUESTION 2.2 â€” MISE Ã€ JOUR SÃ‰QUENTIELLE")
print("="*80)

prior_2 = posterior
likelihood_defaut_2 = evenements['Decouvert >500â‚¬']['P(E|Defaut)']
likelihood_non_defaut_2 = evenements['Decouvert >500â‚¬']['P(E|Non-defaut)']

p_decouvert = likelihood_defaut_2 * prior_2 + likelihood_non_defaut_2 * (1 - prior_2)
posterior_2 = (likelihood_defaut_2 * prior_2) / p_decouvert

print(f"AprÃ¨s Retard : {posterior:.2%}")
print(f"AprÃ¨s DÃ©couvert : {posterior_2:.2%}")

# Graphique Ã©volution
fig, ax = plt.subplots(figsize=(8, 5))

etapes = ['Prior initial', 'AprÃ¨s Retard', 'AprÃ¨s DÃ©couvert']
probas = [prior * 100, posterior * 100, posterior_2 * 100]

ax.plot(etapes, probas, marker='o')
ax.set_ylabel("ProbabilitÃ© dÃ©faut (%)")
ax.set_title("Mise Ã  jour sÃ©quentielle - ThÃ©orÃ¨me de Bayes")

plt.show()

# =============================================================================
# QUESTION 2.3 â€” FONCTION GÃ‰NÃ‰RIQUE BAYES
# =============================================================================

print("\n" + "="*80)
print("QUESTION 2.3 â€” FONCTION GÃ‰NÃ‰RIQUE BAYES")
print("="*80)

def bayes_update(prior, likelihood_pos, likelihood_neg):
    """
    Mise Ã  jour Bayes:
    P(A|B) = P(B|A) P(A) / P(B)
    """

    if not (0 <= prior <= 1):
        raise ValueError("prior doit Ãªtre dans [0,1]")
    if not (0 <= likelihood_pos <= 1):
        raise ValueError("likelihood_pos doit Ãªtre dans [0,1]")
    if not (0 <= likelihood_neg <= 1):
        raise ValueError("likelihood_neg doit Ãªtre dans [0,1]")

    p_evidence = likelihood_pos * prior + likelihood_neg * (1 - prior)

    if p_evidence == 0:
        return 0.0

    return (likelihood_pos * prior) / p_evidence


print("âœ“ Fonction bayes_update crÃ©Ã©e")

# Test client segment Risque
prior_risque = segments['Risque']['taux_defaut']
post_1 = bayes_update(prior_risque,
                      evenements['Retard paiement']['P(E|Defaut)'],
                      evenements['Retard paiement']['P(E|Non-defaut)'])

post_2 = bayes_update(post_1,
                      evenements['Decouvert >500â‚¬']['P(E|Defaut)'],
                      evenements['Decouvert >500â‚¬']['P(E|Non-defaut)'])

post_3 = bayes_update(post_2,
                      evenements['Refus credit ailleurs']['P(E|Defaut)'],
                      evenements['Refus credit ailleurs']['P(E|Non-defaut)'])

print(f"\nSegment Risque - Prior : {prior_risque:.2%}")
print(f"AprÃ¨s 3 Ã©vÃ©nements : {post_3:.2%}")

# =============================================================================
# QUESTION 2.4 â€” MATRICE CONFUSION
# =============================================================================

print("\n" + "="*80)
print("QUESTION 2.4 â€” MATRICE CONFUSION")
print("="*80)

n_total = 10000
n_defauts_reels = 500
n_non_defauts_reels = n_total - n_defauts_reels

tp = 400
fp = 950
fn = n_defauts_reels - tp
tn = n_non_defauts_reels - fp

precision = tp / (tp + fp)

print(f"Precision : {precision:.2%}")
print(f"P(DÃ©faut|Retard) Bayes : {posterior:.2%}")
print(f"DiffÃ©rence : {abs(precision - posterior):.4f}")

print("\nâœ“ FIN PARTIE 2")

# =============================================================================
# TP PARTIE 3 â€” K-NEAREST NEIGHBORS (KNN) ET Ã‰VALUATION MODÃˆLE
# Classification crÃ©dit et optimisation hyperparamÃ¨tres
# =============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    roc_curve,
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score,
    accuracy_score,
    precision_recall_curve
)

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 5)

print("="*80)
print("TP PARTIE 3 â€” KNN ET Ã‰VALUATION MODÃˆLE")
print("="*80)

# =============================================================================
# 1. GÃ‰NÃ‰RATION DATASET SYNTHÃ‰TIQUE CRÃ‰DIT
# =============================================================================

np.random.seed(42)
n = 2000

data = pd.DataFrame({
    "revenu": np.random.normal(3000, 800, n),
    "age": np.random.normal(40, 10, n),
    "decouvert_moyen": np.random.normal(200, 150, n),
    "retards": np.random.poisson(1.5, n)
})

# CrÃ©ation probabilitÃ© dÃ©faut artificielle
prob_defaut = (
    0.002 * data["decouvert_moyen"] +
    0.15 * data["retards"] -
    0.0003 * data["revenu"]
)

prob_defaut = 1 / (1 + np.exp(-prob_defaut))
data["defaut"] = np.random.binomial(1, prob_defaut)

print("\nAperÃ§u dataset :")
print(data.head())

# =============================================================================
# 2. SPLIT TRAIN / TEST
# =============================================================================

X = data.drop("defaut", axis=1)
y = data["defaut"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# Standardisation
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =============================================================================
# 3. MODÃˆLE KNN
# =============================================================================

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

y_pred = knn.predict(X_test_scaled)
y_proba = knn.predict_proba(X_test_scaled)[:, 1]

print("\nClassification Report :")
print(classification_report(y_test, y_pred))

# =============================================================================
# 4. MÃ‰TRIQUES
# =============================================================================

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_proba)

print("\nMÃ‰TRIQUES :")
print(f"Accuracy  : {accuracy:.3f}")
print(f"Precision : {precision:.3f}")
print(f"Recall    : {recall:.3f}")
print(f"F1-score  : {f1:.3f}")
print(f"ROC-AUC   : {roc_auc:.3f}")

# =============================================================================
# 5. MATRICE DE CONFUSION
# =============================================================================

cm = confusion_matrix(y_test, y_pred)

plt.figure()
sns.heatmap(cm, annot=True, fmt="d", cmap="Reds")
plt.title("Matrice de confusion")
plt.xlabel("PrÃ©dit")
plt.ylabel("RÃ©el")
plt.show()

# =============================================================================
# 6. COURBE ROC
# =============================================================================

fpr, tpr, _ = roc_curve(y_test, y_proba)

plt.figure()
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.title("Courbe ROC")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()

# =============================================================================
# 7. OPTIMISATION DU PARAMÃˆTRE K (Cross-validation)
# =============================================================================

print("\nOptimisation du nombre de voisins...")

k_range = range(1, 21)
cv_scores = []

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train_scaled, y_train,
                             cv=cv, scoring="roc_auc")
    cv_scores.append(scores.mean())

best_k = k_range[np.argmax(cv_scores)]

print(f"Meilleur k selon ROC-AUC : {best_k}")

plt.figure()
plt.plot(k_range, cv_scores, marker='o')
plt.title("Optimisation de k (Cross-Validation ROC-AUC)")
plt.xlabel("Nombre de voisins (k)")
plt.ylabel("ROC-AUC")
plt.show()

print("\nâœ“ FIN PARTIE 3")



```
